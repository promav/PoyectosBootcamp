# FUNCIONES OUTLIERS

## TUKEY

```python
def tukey(array, k=1.5):
    q1 = np.quantile(array, q=0.25, method="lower")
    q3 = np.quantile(array, q=0.75, method="lower")
    iqr = q3 - q1
    lim_min = q1 - k*iqr
    lim_max = q3 + k*iqr
    return lim_min, lim_max
    
    
```

## Z-SCORE

```python
def z_score(array, z=3):
    mean = np.mean(array)
    std = np.std(array)
    lim_min = mean - z*std
    lim_max = mean + z*std
    return lim_min, lim_max
```

# REPRESENTACION DE LIMITES EN PRECIOS

```python
tukey_min, tukey_max = tukey(prices)
z_min, z_max = z_score(prices)

plt.hist(prices, bins=25, color='green')
plt.axvline(tukey_min, color='red', linestyle="dashed", label =f'tukey_min: {tukey_min:.2f}')
plt.axvline(tukey_max, color='red', linestyle="dashed", label =f'tukey_max: {tukey_max:.2f}')
plt.axvline(z_min, color='blue', linestyle="dashed", label =f'z_min: {z_min:.2f}')
plt.axvline(z_max, color='blue', linestyle="dashed", label =f'z_max: {z_max:.2f}')
plt.legend()
```

# MATRIZ DE CORRELACION

```python
columnas = [prices, bedrooms, bathrooms, m2]

def matriz_correlacion(array):
    matriz = []
    for row in array:
        fila = []
        for col in array:
            fila.append(stats.pearsonr(row, col)[0])
        matriz.append(fila)
    return np.array(matriz)

matriz_idealista = matriz_correlacion(columnas)

title = ['price', 'bedrooms', 'bathrooms', 'm2']
sns.heatmap(matriz_idealista, annot=True, cmap='viridis', xticklabels=title, yticklabels=title)
matriz_idealista
```

# ESTANDARIZACION

```python
def estandarizacion(array):
    mean = np.mean(array)
    std = np.std(array)
    return np.array([((i - mean)/std) for i in array])

prices_std = estandarizacion(prices)
bedrooms_std = estandarizacion(bedrooms)
bathrooms_std = estandarizacion(bathrooms)
m2_std = estandarizacion(m2)

plt.hist(prices_std, bins=25)
print(f"Media estandarizada: {np.mean(prices_std):.2f}")
print(f"Mediana estandarizada: {np.median(prices_std):.2f}")
print(f"Desviacion estandarizada: {np.std(prices_std):.2f}")
```

# HIPOTESIS

```python
precios_mas_3baños = prices[bathrooms >= 3]
precios_1_2baños = prices[(bathrooms == 1) | (bathrooms == 2)]
# nivel de significancia para aceptar o rechazar pruebas
alpha = 0.05

_, p_3baños = stats.normaltest(precios_mas_3baños)
_, p_1_2baños = stats.normaltest(precios_1_2baños)

print("pruebas de normalidad")
print(f"baños >= 3 {p_3baños} y 1-2 baños {p_1_2baños}")

_, p_var = stats.levene(precios_mas_3baños, precios_1_2baños)
print("prueba de varianza")
print(f"p_var {p_var}")

suposiciones = p_3baños > alpha and p_1_2baños > alpha and p_var > alpha
if suposiciones:
    print("Se cumple la asunción de NORMALIDAD Y HOMOGENEIDAD DE VARIANZAS")
else:
    print("NO se cumple la asunción de NORMALIDAD Y HOMOGENEIDAD DE VARIANZAS") 
```

```python
_, p = stats.mannwhitneyu(precios_mas_3baños, precios_1_2baños, alternative="greater")
print(f"Contraste de hipótesis Mann-Whitney p-value: {p}")

if p < alpha:
    print("H1 VERDADERO (H alternativa): Las casas con 1-2baños son mas caras. RECHAZAMOS H0 (H Nula).")
else:
    print("H0 VERDADERO (H nula): Las casas con 3 baños son mas caras. RECHAZAMOS H1 (H alternativa).")
```
